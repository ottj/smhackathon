```python
import pandas

songs = pandas.read_csv('dataset.csv')
```

```python
from gensim.models import KeyedVectors

word_vectors = KeyedVectors.load_word2vec_format('~/Downloads/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
import numpy as np


def cos_sim(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    return dot_product / (norm_a * norm_b)


def text_vector(text):
    text_vectors = [word_vectors[word] for word in text.split() if word in word_vectors.vocab]

    if len(text_vectors) > 0:
        return np.average(text_vectors, axis=0)
    else:
        return None


songs['vector'] = songs['text'].apply(lambda lyrics: text_vector(lyrics))
songs = songs[songs['vector'].notnull()] 
```

```python
from IPython.core.display import display, HTML


def show_lyrics(index):
    display(HTML(songs.loc[index, 'text']))
    

def search(text):
    ref_vector = text_vector(text)
    sorted_songs = songs.copy()
    sorted_songs["score"] = sorted_songs["vector"].apply(lambda vector: cos_sim(ref_vector, vector))
    return sorted_songs.sort_values('score', ascending=False)
```
